```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Generar una serie temporal artificial (por ejemplo, una señal senoidal con ruido)
np.random.seed(42)
n = 500
time = np.arange(n)
series = 10 + 0.5 * time + np.sin(time) + 2*np.random.normal(0, 1, n) + 15*np.sin(time)

# Crear un DataFrame de pandas
data = pd.DataFrame({'Time': time, 'Value': series})
data.set_index('Time', inplace=True)

# Graficar la serie temporal
plt.figure(figsize=(10, 6))
plt.plot(data, label='Serie Temporal')
plt.title('Serie Temporal Artificial')
plt.show()

# Dividir los datos en entrenamiento y prueba
train_size = int(len(series) * 0.8)
train, test = series[:train_size], series[train_size:]

# Definir y entrenar el modelo ARIMA (p=5, d=1, q=0 como ejemplo)
model = ARIMA(train, order=(5, 2, 1))
model_fit = model.fit()

# Predecir los valores en el conjunto de prueba con intervalos de confianza
forecast_obj = model_fit.get_forecast(steps=len(test))
forecast = forecast_obj.predicted_mean
conf_int = forecast_obj.conf_int()

# Graficar los resultados junto con los intervalos de confianza (zona de error)
plt.figure(figsize=(10, 6))
plt.plot(train, label='Entrenamiento')
plt.plot(np.arange(train_size, n), test, label='Prueba', color='orange')
plt.plot(np.arange(train_size, n), forecast, label='Predicción', color='green')

# Graficar la zona de error (intervalo de confianza)
plt.fill_between(np.arange(train_size, n), 
                 conf_int[:, 0],  # Límite inferior del intervalo de confianza
                 conf_int[:, 1],  # Límite superior del intervalo de confianza
                 color='green', alpha=0.3, label='Intervalo de confianza')

plt.title('Predicción de Serie Temporal con ARIMA (con Zona de Error)')
plt.legend()
plt.show()

# Mostrar las predicciones y los intervalos de confianza
print(forecast)
print(conf_int)
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Crear datos iniciales
initial_data = np.random.randn(100)

# Crear y ajustar el modelo ARIMA inicial
model = ARIMA(initial_data, order=(5, 1, 0))
model_fit = model.fit()

# Predicciones iniciales
print("Predicción inicial:")
print(model_fit.forecast(steps=5))
```

```python
# Nuevos datos que quieres añadir al modelo
new_data = np.random.randn(10)

# Extender el modelo con los nuevos datos
model_fit = model_fit.append(new_data[0:1], refit=False)

# Predicciones después de actualizar con nuevos datos
print("Predicción después de añadir nuevos datos:")
print(model_fit.forecast(steps=5))
```
